```python
# расчет моды вручную (пример для одного столбца Age)
moda = {}
for line in horse_analitic.Age.round():
    if line not in moda:
        moda[line] = 1
    else:
        moda[line] += 1
        
max_line = 0
mode = None
for k, v in moda.items():
    if max_line <v:
        max_line = v
        mode = k
print('Значение моды "Age":', mode)
```

# Выбросы

```python
# определим выбросы для количественных данных, Rectal_temperature
# доверительный интервал (межквартильный размах)
q1 = horse_analitic.Rectal_temperature.quantile(0.25)
q3 = horse_analitic.Rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_analitic[horse_analitic.Rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('Rectal_temperature')
remove_outliers
```

```python
# определяем выбросы по Rectal_temperature
emissions_Rectal_temperature = pd.concat([horse_analitic, remove_outliers]).drop_duplicates(keep=False)
emissions_Rectal_temperature.loc[~pd.isnull(emissions_Rectal_temperature.Rectal_temperature), :].sort_values('Rectal_temperature', ascending=False)
```

```python
# есть ли схожие данные для замены пропуска строки 141?
a = horse_analitic.loc[((horse_analitic.Mucous_embranes==5.0)&(horse_analitic.Pulse.between(35.0, 52.0, inclusive=True))), :]
a.Temperature_of_extremities.mean()
a.Pain.mean()
a.Abdominal_distension.mean()
a
```

```python
# сделаем замену
dt = horse_analitic.iloc[[141], :]
dt.Temperature_of_extremities.fillna(a.Temperature_of_extremities.mean(), inplace=True)
dt.Pain.fillna(a.Pain.mean(), inplace=True)
dt.Abdominal_distension.fillna(a.Abdominal_distension.mean(), inplace=True)
horse_analitic.iloc[[141], :]=dt
horse_analitic.iloc[[141], :]
```

выбросы по Rectal_temperature кажутся естественными (оставляем),

по другим показателям тоже идут не нормальные данные - 

либо лошадь болеет, испытывает боль, либо перегрелась...

потеря данных одной лошади из данного выброса - можно заполнить пропуски по аналогии со схожими симптомами

```python
# определим выбросы для количественных данных, Pulse
q1_2 = horse_analitic.Pulse.quantile(0.25)
q3_2 = horse_analitic.Pulse.quantile(0.75)
iqr_2 = q3_2 - q1_2
lower_bound_2 = q1_2 - (1.5 * iqr_2)
upper_bound_2 = q3_2 + (1.5 * iqr_2)
remove_outliers_2 = horse_analitic[horse_analitic.Pulse.between(lower_bound_2, upper_bound_2, inclusive=True)].sort_values('Pulse')
remove_outliers_2
```

```python
# определяем выбросы по Pulse
emissions_Pulse = pd.concat([horse_analitic, remove_outliers_2]).drop_duplicates(keep=False)
emissions_Pulse.loc[~pd.isnull(emissions_Pulse.Pulse), :].sort_values('Pulse', ascending=False)
```

```python
# есть ли схожие данные для замены пропуска строки 41?
a2 = horse_analitic.loc[horse_analitic.Pulse==150.0, :]
a2
```

```python
# сделаем замену
dt2 = horse_analitic.iloc[[41], :]
dt2.Temperature_of_extremities.fillna(a2.Temperature_of_extremities.mean(), inplace=True)
dt2.Mucous_embranes.fillna(a2.Mucous_embranes.mean(), inplace=True)
dt2.Pain.fillna(a2.Pain.mean(), inplace=True)
dt2.Abdominal_distension.fillna(a2.Abdominal_distension.mean(), inplace=True)
horse_analitic.iloc[[41], :]=dt2
horse_analitic.iloc[[41], :]
```

выбросы по Pulse кажутся естественными (оставляем),

по другим показателям тоже идут не нормальные данные -

```python
# пропуски в столбце Rectal_temperature
horse_analitic.loc[pd.isnull(horse_analitic.Rectal_temperature), :]
```

```python
# поскольку распределение температур выглядит относительно симметрично, \
# для замены можно взять или моду, или медиану, или среднее
# заменим пропуски столбца Rectal_temperature медианой
b = horse_analitic.loc[pd.isnull(horse_analitic.Rectal_temperature), :]
b.Rectal_temperature.fillna(horse_analitic.Rectal_temperature.median(), inplace=True)
horse_analitic.loc[pd.isnull(horse_analitic.Rectal_temperature), :] = b
horse_analitic.loc[pd.isnull(horse_analitic.Rectal_temperature), :]
```

```python
horse_analitic.Rectal_temperature.isnull().sum()
```

```python
# пропуски в столбце Pulse
horse_analitic.loc[pd.isnull(horse_analitic.Pulse), :]
```

```python
# пропуски в столбце Pulse можно заполнить медианой группировки
# print(horse_analitic.groupby('Rectal_temperature').Pulse.median())
horse_analitic.Pulse.fillna(horse_analitic.groupby('Rectal_temperature').Pulse.transform('median'), inplace=True)
horse_analitic.Pulse.isnull().sum()
```

```python
# последние четыре столбца качественных характеристик заменим модой
horse_analitic.Temperature_of_extremities.fillna(horse_analitic.Temperature_of_extremities.mode()[0], inplace=True)
horse_analitic.Temperature_of_extremities.isnull().sum()
```

```python
horse_analitic.Mucous_embranes.fillna(horse_analitic.Mucous_embranes.mode()[0], inplace=True)
horse_analitic.Mucous_embranes.isnull().sum()
```

```python
horse_analitic.Pain.fillna(horse_analitic.Pain.mode()[0], inplace=True)
horse_analitic.Pain.isnull().sum()
```

```python
horse_analitic.Abdominal_distension.fillna(horse_analitic.Abdominal_distension.mode()[0], inplace=True)
horse_analitic.Abdominal_distension.isnull().sum()
```

```python
# удалим строку с пропуском в столбце Surgery (осталась одна) 
horse_analitic.dropna(inplace=True)
```

```python
# проверяем датафрейм, он должен получиться без пропусков
horse_analitic.isnull().sum()
```
