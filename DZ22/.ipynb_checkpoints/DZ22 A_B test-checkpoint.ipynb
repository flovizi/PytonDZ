## Проверка гипозез. А\В тестирование ##

```python
import numpy as np
import pandas as pd
import scipy.stats as stats 
import matplotlib.pyplot as plt
```

<div class="alert alert-info">
<b>Задание № 1</b>
Вам даны результаты проверки двух препаратов.
    
Требуется:
- Оценить, есть ли различие в эффекте у этих пациентов?
- Визуально подтвердить результат
    
`A` - результаты для группы людей получающих препарат А.

`B` - результаты для другой группы людей, получающих препарат B.
    
</div>

```python
np.random.seed(11)
A = stats.norm.rvs(scale=30, loc=11, size=100)
B = A + stats.norm.rvs(scale=54, loc=11, size=100)

# группы А, В - независимые, применим t-тест Стьюдента (только две группы)

stat, p = stats.ttest_ind(A, B)
print(f'Statistics = {stat:.3f}, {p:.3f}')

alfa = 0.05 # допустим
if p > alfa:
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
    
#plt.hist(pd.DataFrame(A), bins = 20);
#plt.hist(pd.DataFrame(B), bins = 20);

pd.DataFrame(A).hist(bins = 20);
pd.DataFrame(B).hist(bins = 20);
```

<div class="alert alert-info">
<b>Задание № 2</b>
Вам даны результаты проверки двух препаратов. 
    
Требуется: 
- Оценить, есть ли различие в эффекте у этих пациентов?
- Визуально подтвердить результат
    
`А` - группа людей без препарата.

`B` - та же группа людей, но получающая препарат
</div>

```python
np.random.seed(11)
A = stats.norm.rvs(scale=30, loc=11, size=100)
B = A + stats.norm.rvs(scale=54, loc=11, size=100)

# группы А, В - зависимые, применим критерий Уилкоксона

from scipy.stats import wilcoxon

stat, p = wilcoxon(A, B)
print(f'Statistics = {stat:.3f}, {p:.3f}')

alfa = 0.05 
if p > alfa:
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
    
#plt.hist(pd.DataFrame(A), bins = 20);
#plt.hist(pd.DataFrame(B), bins = 20);

pd.DataFrame(A).hist(bins = 20);
pd.DataFrame(B).hist(bins = 20);
```

`nutri = [1,2,4,6,5,6,7,5,4,5,6,7,4,3,5,5,6,5,4,3,5]`

`mineral =[2,1,1,3,2,4,2,4,5,4,3,2,3,2,3,1,3,4,5,1,4]`

```python
water = [1,2,3,4,2,4,2,4,5,2,3,4,2,1,3,4,3,2,5,1]
nutri = [1,2,4,6,5,6,7,5,4,5,6,7,4,3,5,5,6,5,4,3,5]
mineral =[2,1,1,3,2,4,2,4,5,4,3,2,3,2,3,1,3,4,5,1,4]

# для множественных сравнений применим дисперсионный анализ

from scipy.stats import f_oneway

F, p = f_oneway(water, nutri, mineral)

alfa = 0.05
print(F, p)

if p > alfa:
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
    
#plt.hist(pd.DataFrame(water), bins = 20);
#plt.hist(pd.DataFrame(nutri), bins = 20);
#plt.hist(pd.DataFrame(mineral), bins = 20);


pd.DataFrame(water).hist(bins = 20);
pd.DataFrame(nutri).hist(bins = 20);
pd.DataFrame(mineral).hist(bins = 20);
```

```python
# видим, что есть различия по трем экспериментам.
# уточним различие между удобрением и минеральной водой

from scipy.stats import f_oneway

F, p = f_oneway(nutri, mineral)

alfa = 0.05
print(F, p)

if p > alfa:
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
    
#plt.hist(pd.DataFrame(nutri), bins = 20);
#plt.hist(pd.DataFrame(mineral), bins = 20);

pd.DataFrame(nutri).hist(bins = 20);
pd.DataFrame(mineral).hist(bins = 20);
```

# Данные A/B тестирования

```python
cookie_cats = 'https://raw.githubusercontent.com/a-milenkin/datasets_for_t-tests/main/cookie_cats.csv'

data = pd.read_csv(cookie_cats, error_bad_lines=False)
data.head()
```

**Данные получены от 90 189 игроков, которые установили игру во время проведения AB-теста. Переменные:**

`userid` - уникальный номер, идентифицирующий каждого игрока.

версия - был ли игрок помещен в контрольную группу (gate_30 - ворота на уровне 30) или в тестовую группу (gate_40 - ворота на уровне 40).

`sum_gamerounds` - количество игровых раундов, сыгранных игроком в течение первой недели после установки

`retention_1` - проигрыватель вернулся и поиграл через 1 день после установки?

`retention_7` - проигрыватель вернулся и играл через 7 дней после установки?

Когда игрок устанавливал игру, ему случайным образом назначали `gate_30` или `gate_40`.

<div class="alert alert-info">
<b>Задание: провести анализ данных и сделать выводы о лучшем месте для установки ворот. </b>
    
Замечание: приведенных вспомогательных фрагментов кода не хватает для ответа на вопрос. 
Делать выводы необходимо опираясь на статистические тесты, разобранные в лекции. Большим плюсом будет построение визуализиция для подтверждения ваших выводом. 
</div>

```python
#Сгрупируем результаты теста по версиям игры.
data.groupby('version').count()
```

```python
# Ваши выводы
# примерно одинаковый объем выборок двух групп
```

```python
# Подсчет количества игроков для каждого количества раундов
plot_df = data.groupby('sum_gamerounds')['userid'].count()
plot_df.hist(bins=15)  # попробуйте поварировать параметр bins
#plt.hist(plot_df, bins=15);
```

```python
# пыталась построить график локтя (думала посмотреть лучший порог)

# увеличим размер графика
from pylab import rcParams
rcParams['figure.figsize'] = 13,6

# строим график локтя

from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression

c = []
a = plot_df.reset_index()
for i, j in a.iterrows():
    c.append([j.sum_gamerounds, j.userid])
cc = np.array(c)


# создаем список для инерции
k_inertia = []
# задаем диапазон кластеров
ks = range(1, 50)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(cc)
    # добавляем инерцию каждой модели в список
    k_inertia.append(clf_kmeans.inertia_)
    
plt.plot(ks, k_inertia);
plt.plot(ks, k_inertia ,'ro');
```

**Ваши выводы**

```python
# количество игроков после определенного раунда (порога) резко снижается
```

```python
# # Creating an list with bootstrapped means for each AB-group
boot_1d = []
for i in range(1000):
    boot_mean = data.sample(frac = 1, replace = True).groupby('version')['retention_7'].mean()  #retention_1
    boot_1d.append(boot_mean)
    
# # Преобразование списка в DataFrame
boot_1d = pd.DataFrame(boot_1d)
    
# # A Kernel Density Estimate plot of the bootstrap distributions
boot_1d.plot(kind='density')
```

```python
# Ваши выводы?
# чаще возвращались играть через 7 дней после установки gate_30
```
