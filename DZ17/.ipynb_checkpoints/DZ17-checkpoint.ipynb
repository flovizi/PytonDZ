```python
table_2015 = pd.read_csv(r'C:\Users\Urmas\Desktop\Jupyter\test\Statistika\Визуализация\Tables for homework\2015.csv', encoding = 'utf-8')
table_2015['Year'] = 2015
table_2015.columns
table_2015.rename(columns={'Dystopia Residual': 'Dystopia_Residual', 'Economy (GDP per Capita)': 'Economy_GDP_per_Capita', 'Happiness Rank': 'Happiness_Rank', 'Happiness Score': 'Happiness_Score', 'Health (Life Expectancy)': 'Health(Life_Expectancy)', 'Standard Error': 'Standard_Error', 'Trust (Government Corruption)': 'Trust(Government_Corruption)'}, inplace=True)
table_2015.head()
```

```python
table_2016 = pd.read_csv(r'C:\Users\Urmas\Desktop\Jupyter\test\Statistika\Визуализация\Tables for homework\2016.csv', encoding = 'utf-8')
table_2016['Year'] = 2016
table_2016.columns
table_2016.rename(columns={'Dystopia Residual': 'Dystopia_Residual', 'Economy (GDP per Capita)': 'Economy_GDP_per_Capita', 'Happiness Rank': 'Happiness_Rank', 'Happiness Score': 'Happiness_Score', 'Health (Life Expectancy)': 'Health(Life_Expectancy)', 'Upper Confidence Interval': 'Upper_Confidence_Interval', 'Lower Confidence Interval': 'Lower_Confidence_Interval', 'Trust (Government Corruption)': 'Trust(Government_Corruption)'}, inplace=True)
table_2016.head()
```

```python
table_2017 = pd.read_csv(r'C:\Users\Urmas\Desktop\Jupyter\test\Statistika\Визуализация\Tables for homework\2017.csv', encoding = 'utf-8')
table_2017['Year'] = 2017
table_2017.columns
table_2017.rename(columns={'Dystopia.Residual': 'Dystopia_Residual', 'Economy..GDP.per.Capita.': 'Economy_GDP_per_Capita', 'Happiness.Rank': 'Happiness_Rank', 'Happiness.Score': 'Happiness_Score', 'Health..Life.Expectancy.': 'Health(Life_Expectancy)', 'Trust..Government.Corruption.': 'Trust(Government_Corruption)'}, inplace=True)
table_2017.head()
```

```python
table_2018 = pd.read_csv(r'C:\Users\Urmas\Desktop\Jupyter\test\Statistika\Визуализация\Tables for homework\2018.csv', encoding = 'utf-8')
table_2018['Year'] = 2018
table_2018.columns
table_2018.rename(columns={'Overall rank': 'Happiness_Rank', 'Country or region': 'Country', 'Score': 'Happiness_Score', 'GDP per capita': 'Economy_GDP_per_Capita',
                           'Social support': 'Family', 'Healthy life expectancy': 'Health(Life_Expectancy)',
                           'Freedom to make life choices': 'Freedom',
                           'Perceptions of corruption': 'Trust(Government_Corruption)'}, inplace=True)
table_2018.head()
```

```python
table_2019 = pd.read_csv(r'C:\Users\Urmas\Desktop\Jupyter\test\Statistika\Визуализация\Tables for homework\2019.csv', encoding = 'utf-8')
table_2019['Year'] = 2019
table_2019.columns
table_2019 = table_2019.rename(columns={'Overall rank':'Happiness_Rank', 'Country or region':'Country',
                                        'Score':'Happiness_Score', 'GDP per capita':'Economy_GDP_per_Capita',
                                        'Healthy life expectancy':'Health(Life_Expectancy)',
                                        'Freedom to make life choices':'Freedom',
                                        'Perceptions of corruption':'Trust(Government_Corruption)',
                                        'Social support':'Family'})
table_2019.head()
```

```python
table_happy = pd.concat([table_2015, table_2016, table_2017, table_2018, table_2019], axis=0, ignore_index=True)
table_happy.head()
```

```python
#table_happy.isnull().sum()
for col in table_happy.columns:
    missing = table_happy[col].isnull().mean()
    print(f'{col} - {missing :.1%}')
```

```python
table_happy.drop(columns=['Standard_Error','Lower_Confidence_Interval','Upper_Confidence_Interval', 'Whisker.high','Whisker.low'],inplace=True)
table_happy.head()
```

```python
# заполним пропуски по регионам по известным парам "страна-регион"
country_miss_region = table_happy.loc[pd.isnull(table_happy.Region), :].Country.unique()
country_miss_region
miss_region = table_happy.loc[pd.isnull(table_happy.Region), :]
miss_region
not_miss_region = table_happy.loc[~pd.isnull(table_happy.Region), :]
not_miss_region
```

```python
# сформируем словарь пар "страна-регион"
country = list(miss_region.Country)
country
d = {}
for x in country:
    y = list(not_miss_region[not_miss_region.Country == x].Region)
    for i in y:
        d[x]=i
d
```

```python
for c in country_miss_region:
    try:
        table_happy['Region'].fillna(d[c], inplace=True)
    except:
        pass

for col in table_happy.columns:
    missing = table_happy[col].isnull().mean()
    print(f'{col} - {missing :.1%}')

# теперь имеем заполненный столбец Region
```

```python
# заполним столбец Dystopia_Residual
# Happiness_Score - это сумма всех столбцов
table_happy = table_happy.apply(pd.to_numeric, errors = 'ignore')  # для достоверности переведем числовые значения в числа
table_happy['Dystopia_Residual'].fillna((table_happy['Happiness_Score'] - table_happy['Economy_GDP_per_Capita'] - table_happy['Family'] - table_happy['Health(Life_Expectancy)'] - table_happy['Freedom'] - table_happy['Trust(Government_Corruption)'] - table_happy['Generosity']), inplace=True)
table_happy.loc[pd.isnull(table_happy.Dystopia_Residual), :]  # остается одна строка с неизвестными данными - удалим ее
table_happy.dropna(inplace=True)
table_happy.isnull().sum() # сформировали датафрейм без NaN
```

```python
table_happy_region_year = table_happy.groupby(['Year', 'Region'])['Happiness_Score'].mean().reset_index()
table_happy_region_year.head()
```

```python
table_happy_general = table_happy_region_year.pivot_table(index = 'Year', columns = 'Region', values = 'Happiness_Score', aggfunc = 'mean', margins=True)
table_happy_general
```

```python
# данные за 2017-2019 гг. представлены только за один регион, тогда будем рассматривать совокупность 2015 и 2016 гг.
table_happy_general = table_happy_general.loc[table_happy_general.index.isin([2015, 2016, 'All'])]
table_happy_general = table_happy_general.apply(pd.to_numeric, errors = 'ignore')
table_happy_general
```

```python
table_happy_general2 = table_happy_general.loc[table_happy_general.index.isin([2015, 2016]), ['Australia and New Zealand', 'Central and Eastern Europe', 
                                                                                               'Eastern Asia', 'Latin America and Caribbean', 
                                                                                               'Middle East and Northern Africa', 'North America', 
                                                                                               'Southeastern Asia', 'Southern Asia', 
                                                                                               'Sub-Saharan Africa', 'Western Europe']]
table_happy_general2                                              
```

```python
for region in table_happy_general2.columns:
    plt.plot(table_happy_general2.index, table_happy_general2[region])
    plt.title('Динамика состояния счастья по регионам и годам')
    plt.xlabel('Год')
    plt.ylabel('Уровень счастья')
    plt.legend(table_happy_general2, loc='lower right')
    # состояние счастья относительно стабильно внутри региона, немного снижается
    # в регионах разный уровень счастья
```

```python
table_happy_general3 = table_happy_general.loc[table_happy_general.index.isin([2015, 2016])]
table_happy_general3
```

```python
plt.plot(table_happy_general3.index, table_happy_general3['All'])
plt.title('Динамика состояния счастья по годам')
plt.xlabel('Год')
plt.ylabel('Уровень счастья')
plt.legend(table_happy_general, loc='lower left')
# наблюдается общее снижение уровня счастья
```

```python
tabler = table_happy.loc[table_happy.Year.isin([2015, 2016])]
tabler.drop(columns=['Year', 'Happiness_Rank'], inplace=True)
tabler = tabler.groupby('Region').mean().reset_index()
tabler
```
