* Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv(r'C:\Users\Urmas\Desktop\Jupyter\test\Pandas продвинутый\DZ\ml-latest-small\ratings.csv')
ratings.head()
```

```python
"""
подсчет времени жизни каждого userId
"""
a = ratings.groupby('userId').apply(lambda row: row.timestamp.max() - row.timestamp.min()).reset_index()
a.columns = ['userId', 'times']
a.head()
```

```python
"""
подсчет и выборка userId с количеством оценок более 100
"""
a2 = ratings['userId'].value_counts().reset_index()
a2.columns = ['userId', 'score']
a2 = a2[a2['score'] > 100]
a2.head()
```

```python
"""
объединение таблиц с выводом искомого среднего по времени жизни userId
"""
joined = a2.merge(a, on= 'userId', how='left')['times'].mean()
joined
```

* Какие дополнительные признаки можно получить из координат? Ведь это просто числа, которые сами по себе мало что дают.

Ответ: к сожалению, не понимаю (не представляю) условия задачи первого и второго пунктов, но, по принципу, думаю так - вынести таблицу данных с координатами в словарь и отдуда извлекать (присоединять) данные. Из координат можно получить анализ по приоритетам покупок относительно региона.

```python
# допустим, имеем таблицу user_id с визитами и покупками
table_vizit0 = {
    'user_id': [1, 2, 3, 4, 5, 6, 7] ,
    'visit': [1, 23, 4, 5, 24,30, 7],
    'custom': [2, 4, 3, 6, 7, 11, 12]
}
table_vizit0
```

```python
# переведем таблицу user_id с визитами и покупками в DataFrame
table_vizit = pd.DataFrame(table_vizit0)
table_vizit
```

```python
# допустим, имеем таблицу user_id с координатами
table_coordinate0 = [
    [1, '56.755831/37.617673'], 
    [2, '55.755832/37.617673'], 
    [3, '56.785831/37.617673'], 
    [4, '55.755832/35.617673'], 
    [5, '56.755835/37.617673'], 
    [6, '58.755832/37.617673'], 
    [7, '56.745835/37.617673']
]
table_coordinate0
```

```python
# переведем таблицу user_id с координатами в словарь
table_coordinate1 = {rec[0]:rec[1] for rec in table_coordinate0}

table_coordinate1
```

```python
# объединяю данные
for i in table_coordinate1:
    table_vizit.loc[table_vizit.user_id == i, 'coordinate'] = table_coordinate1[i]
table_vizit
```

```python
# объединяю данные, если нужно, к примеру, первые две цифры
import re

for i in table_coordinate1:
    table_vizit.loc[table_vizit.user_id == i, 'coordinate'] = str(re.findall(r'^\w{2}' ,table_coordinate1[i]))
table_vizit
```
