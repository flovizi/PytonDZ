```python
# переименуем столбцы из инфо в доп.таблице
list_ = {
            '1': 'Surgery',
            '2': 'Age',
            '3': 'Hospital_Number',
            '4': 'Rectal_temperature',
            '5': 'Pulse',
            '6': 'Respiratory_rate', 
            '7': 'Temperature_of_extremities',
            '8': 'Peripheral_pulse',
            '9': 'Mucous _embranes',
            '10': 'Capillary_refill_time',
            '11': 'Pain',
            '12': 'Peristalsis',
            '13': 'Abdominal_distension',
            '14': 'Nasogastric_tube',
            '15': 'Nasogastric_reflux',
            '16': 'Nasogastric_reflux_PH',
            '17': 'Rectal_examination_feces',
            '18': 'Abdomen',
            '19': 'Packed_cell_volume',
            '20': 'Total_protein',
            '21': 'Abdominocentesis_appearance',
            '22': 'Abdomcentesis_total_protein',
            '23': 'Outcome',
            '24': 'Surgical_lesion?',
            '25': 'Type_of_lesion1',
            '26': 'Type_of_lesion2',
            '27': 'Type_of_lesion3',
            '28': 'Cp_data'
}
names = list(list_.values())
print(names, sep=',')
horse.columns = [names]
horse.head()
```

```python
# какие колонки оставим для анализа
# № 1, 2, 4, 5, 7, 9, 11, 13
num = [1, 2, 4, 5, 7, 9, 11, 13]

column_index = list(map(lambda x: x-1, num))
print(column_index)

# для инфо - названия искомых столбцов
column_names= list(filter(None, map(lambda x: x if names.index(x) in list(map(lambda x: x-1, num)) else None, names)))
print(column_names)
```

```python
# сформируем DataFrame из 8-ми столбцов
# третий и четвертый столбец - количественные непрерывные величины
# первый и одиннадцатый  столбец - качественные номинальные значения (ранжирование незначимо)
# второй, пятый, шестой и тринадцатый столбцы - качественные порядковые значения (ранжирование значимо)

horse_analitic = horse.iloc[:, column_index]
horse_analitic.columns = [column_names]
horse_analitic.head()
```

```python
horse_analitic = horse_analitic.apply(pd.to_numeric)
horse_analitic.info()
```

```python
horse_analitic = horse_analitic.apply(lambda column: pd.to_numeric(column), axis=0)
horse_analitic.info()
```

```python
horse_analitic.describe()
```

```python
moda_Age = horse_analitic.Age.mode()
moda_Age
```

```python
# расчет моды для всего DataFrame
moda = horse_analitic.round().mode()
moda
```

```python
# расчет моды вручную
moda = {}
for line in horse_analitic.Age.round():
    if line not in moda:
        moda[line] = 1
    else:
        moda[line] += 1
        
max_line = 0
mode = None
for k, v in moda.items():
    if max_line <v:
        max_line = v
        mode = k
print('Значение моды', moda)
```

# Выбросы

```python
q1 = horse_analitic.Rectal_temperature.quantile(0.25)
q3 = horse_analitic.Rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_analitic[horse_analitic.Rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('Rectal_temperature')
remove_outliers
```

```python
q1_2 = horse_analitic.Pulse.quantile(0.25)
q3_2 = horse_analitic.Pulse.quantile(0.75)
iqr_2 = q3_2 - q1_2
lower_bound_2 = q1_2 - (1.5 * iqr_2)
upper_bound_2 = q3_2 + (1.5 * iqr_2)
remove_outliers_2 = horse_analitic[horse_analitic.Pulse.between(lower_bound_2, upper_bound_2, inclusive=True)].sort_values('Pulse')
remove_outliers_2
```

```python
# определяем выброс
pd.concat([horse_analitic, remove_outliers, remove_outliers_2]).drop_duplicates(keep=False)
```
